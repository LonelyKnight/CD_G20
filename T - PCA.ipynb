```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ds_functions as ds
import g20_functions as g20

data = g20.loadToxic()

data.pop(1024)

variables = data.columns.values
```

```python
from sklearn.decomposition import PCA

cov_mtx = data.cov()
eigvals, eigvecs = np.linalg.eig(cov_mtx)

pca = PCA()
pca.fit(data)
PC = pca.components_
var = pca.explained_variance_

```

```python
# PLOT EXPLAINED VARIANCE RATIO
fig = plt.figure(figsize=(4, 4))
plt.title('Explained variance ratio')
plt.xlabel('PC')
plt.ylabel('ratio')
x_values = [str(i) for i in range(1, len(pca.components_) + 1)]
bwidth = 0.5
ax = plt.gca()
ax.set_xticklabels(x_values)
ax.set_ylim(0.0, 1.0)
ax.bar(x_values[:10], pca.explained_variance_ratio_[:10], width=bwidth)
ax.plot(pca.explained_variance_ratio_[:10])
for i, v in enumerate(pca.explained_variance_ratio_[:10]):
    ax.text(i, v+0.05, f'{v*100:.1f}', ha='center', fontweight='bold')
plt.show()
```

```python
sum_variance = 0
i = -1
while sum_variance < 0.8:
    i += 1
    sum_variance += pca.explained_variance_ratio_[i]
print(i)
```
