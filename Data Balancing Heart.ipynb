```python
import g20_functions as g20


values = g20.balanceData(g20.loadHeart(), dataset = 'Heart')
```

```python
import g20_KNN
import numpy as np
import pandas as pd
for k,v in values.items():
    if k == 'SMOTE':
        X,y = v
        labels = pd.unique(y)
    else:
        X,y,labels = g20.xySplit(v,target='DEATH_EVENT')
    g20_KNN.crossValKNN(X,y,labels,"Heart_"+k)
```
